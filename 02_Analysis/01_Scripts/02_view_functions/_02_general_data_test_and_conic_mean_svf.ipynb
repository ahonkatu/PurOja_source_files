```python
# Initialize a dictionary to store conic means
means = {}

# Loop through files from SVF_1.npy to SVF_22.npy
for i in range(1, 23):
    file_name = f"SVF_{i}.npy"
    file_path = os.path.join(features_path, file_name)
    
    if os.path.exists(file_path):
        # Load the SVF numpy array
        svf_data = np.load(file_path)
        
        # Remove NaN or invalid values (e.g., masked regions)
        valid_data = svf_data[~np.isnan(svf_data)]
        
        # Calculate the conic mean for this file
        conic_mean = np.mean(valid_data)
        
        # Store the result in the dictionary
        means[file_name] = conic_mean
    else:
        print(f"File not found: {file_name}")

# Sort the means dictionary by the conic mean values (from smallest to biggest)
sorted_means = {k: v for k, v in sorted(means.items(), key=lambda item: item[1])}

# Print sorted results
for file_name, mean in sorted_means.items():
    print(f"Mean for {file_name}: {mean}")
```

```python
# Function to check if a NumPy array is a boolean mask with 0s and 1s
def is_boolean_mask(array):
    unique_values = np.unique(array)
    return set(unique_values).issubset({0, 1})

# Loop through Label3m_{i}.npy files from 1 to 21
for i in range(1, 22):
    file_name = f"Label3m_{i}.npy"
    file_path = os.path.join(features_path, file_name)
    
    if os.path.exists(file_path):
        # Load the numpy file
        label_data = np.load(file_path)
        
        # Check if it is a boolean mask
        if is_boolean_mask(label_data):
            print(f"{file_name}: ✅ Boolean mask (only 0s and 1s)")
        # Plot the first valid boolean mask (e.g., Label3m_1.npy)
            if i == 1:  # Plot only Label3m_1.npy
                plt.figure(figsize=(8, 8))
                plt.imshow(label_data, cmap='gray')
                plt.title(f"Visualization of {file_name}")
                plt.colorbar(label="Pixel Values (0 or 1)")
                plt.axis('off')
                plt.show()
        else:
            print(f"{file_name}: ❌ Not a boolean mask (contains values other than 0 and 1)")
    else:
        print(f"{file_name}: File not found")
```

```python
# Apply the nopython JIT optimization for performance
@jit(nopython=True)
def _create_conic_mask(radius, direction):
    """
    Create a conic mask for a direction with a certain radius.
    """
    kernel = np.zeros((2 * radius + 1, 2 * radius + 1), dtype=np.float64)
    coords = np.arange(-radius, radius + 1, dtype=np.int32)  # Ensure dtype is int32
    y = np.zeros((2 * radius + 1, 2 * radius + 1), dtype=np.int32)
    x = np.zeros((2 * radius + 1, 2 * radius + 1), dtype=np.int32)

    for i in range(2 * radius + 1):
        for j in range(2 * radius + 1):
            y[i, j] = coords[i]
            x[i, j] = coords[j]

    # Create a mask for each direction
    if direction == 0:  # topright
        mask = np.logical_and(np.greater(x, y), np.less(x, np.abs(y)))
        mask = np.logical_and(mask, np.less_equal(x**2 + y**2, radius**2))
        mask = np.logical_and(mask, np.greater(x, 0))

    elif direction == 1:  # righttop
        mask = np.logical_and(np.greater(x, np.abs(y)), np.less_equal(x**2 + y**2, radius**2))
        mask = np.logical_and(mask, np.less(y, 0))

    elif direction == 2:  # rightbottom
        mask = np.logical_and(np.greater(x, np.abs(y)), np.less_equal(x**2 + y**2, radius**2))
        mask = np.logical_and(mask, np.greater(y, 0))

    elif direction == 3:  # bottomright
        mask = np.logical_and(np.less(np.abs(x), y), np.less_equal(x**2 + y**2, radius**2))
        mask = np.logical_and(mask, np.greater(x, 0))

    elif direction == 4:  # bottomleft
        mask = np.logical_and(np.less(np.abs(x), y), np.less_equal(x**2 + y**2, radius**2))
        mask = np.logical_and(mask, np.less(x, 0))

    elif direction == 5:  # leftbottom
        mask = np.logical_and(np.greater(np.abs(x), np.abs(y)), np.less(x, np.abs(y)))
        mask = np.logical_and(mask, np.less_equal(x**2 + y**2, radius**2))
        mask = np.logical_and(mask, np.greater(y, 0))

    # Manually assign values based on the mask (avoid using `kernel[mask] = 1`)
    for i in range(kernel.shape[0]):
        for j in range(kernel.shape[1]):
            if mask[i, j]:
                kernel[i, j] = 1

    return kernel

@jit(nopython=True)
def _mean_from_masks(arr, position, masks):
    i, j = position
    radius = (masks.shape[1] - 1) // 2
    values = []

    for mask in masks:
        masked_sum = 0
        masked_count = 0

        for di in range(-radius, radius + 1):
            for dj in range(-radius, radius + 1):
                ni, nj = i + di, j + dj
                if 0 <= ni < arr.shape[0] and 0 <= nj < arr.shape[1]:
                    if mask[di + radius, dj + radius] == 1:
                        masked_sum += arr[ni, nj]
                        masked_count += 1

        mean_value = masked_sum / masked_count if masked_count > 0 else np.inf
        values.append(mean_value)

    return values

#@jit(nopython=True, debug=True)
def conic_mean(arr, mask_radius=6, threshold=0.2):
    # Function implementation
    """
    Applies directional mean calculation with a conic mask to the array.
    """
    masks = np.zeros((6, 2 * mask_radius + 1, 2 * mask_radius + 1), dtype=np.float64)
    for direction in range(6):
        masks[direction] = _create_conic_mask(mask_radius, direction)

    new_arr = arr.copy()

    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            values = _mean_from_masks(arr, (i, j), masks)

            lowest = min(values)

            if lowest < threshold:
                new_arr[i, j] = 0.9894350171089172 * lowest if lowest * 0.9894350171089172 < arr[i, j] else arr[i, j]

    return new_arr

# Function to load raw data and process it
def load_raw_and_create_df(zone_numbers=range(9, 10), chunk_size=(800, 800)):
    features_path = "../../01_Data/01_Raw/features/features"
    
    # Initialize empty DataFrame to store results
    data_frame = pd.DataFrame(columns=["zone_number", "original_skyview_shape", "conic_mean_skyview_shape"])
    
    for zone_number in zone_numbers:
        # Load the skyview array from the corresponding file
        skyview = np.load(f"{features_path}/SVF_{zone_number}.npy").astype(np.float64) #numba is case sensitive
        
        # Print the shape of the skyview array for debugging
        print(f"Zone {zone_number}: Shape of original skyview: {skyview.shape}")
        
        # Apply the conic mean function to the skyview array
      # Inside the `load_raw_and_create_df` function, before calling conic_mean
        skyview_conic_mean = conic_mean(skyview, 4, 0.9894350171089172)
        # # Filter values between 0.9 and 0.96
        # filtered_conic_mean = np.where((skyview_conic_mean >= 0.9) & (skyview_conic_mean <= 0.96), skyview_conic_mean, np.nan)

        # # Plotting the original skyview array and the filtered conic mean processed array
        # plt.figure(figsize=(10, 8))

        # # Plot the original skyview array
        # plt.subplot(1, 2, 1)
        # plt.imshow(skyview, cmap='viridis', origin='lower')
        # plt.title(f"Original Skyview - Zone {zone_number}")
        # plt.colorbar(label="Skyview Value")

        # # Plot the filtered conic mean processed array
        # plt.subplot(1, 2, 2)
        # plt.imshow(filtered_conic_mean, cmap='viridis', origin='lower', vmin=0.9, vmax=0.96)
        # plt.title(f"Filtered Conic Mean (0.9 - 0.96) - Zone {zone_number}")
        # plt.colorbar(label="Filtered Value")

        # # Display the plots
        # plt.tight_layout()
        # plt.show()

        # Now you can reshape the result outside the numba-compiled function
        flattened_skyview_conic_mean = skyview_conic_mean.flatten() 
        
        print(f"Zone {zone_number}: Shape of conic mean processed skyview: {skyview_conic_mean.shape}")
        print(f"Zone {zone_number}: Flattened conic mean shape: {flattened_skyview_conic_mean.shape}")
        
        data_frame = pd.concat([data_frame, pd.DataFrame([{
            "zone_number": zone_number,
            "original_skyview_shape": skyview.shape,
            "conic_mean_skyview_shape": skyview_conic_mean.shape
            }])], ignore_index=True)
        
        # Plotting the original skyview array and the conic mean processed array
        plt.figure(figsize=(10, 8))

        # Plot the original skyview array
        plt.subplot(1, 2, 1)
        plt.imshow(skyview, cmap='viridis', origin='lower')
        plt.title(f"Original Skyview - Zone {zone_number}")
        plt.colorbar(label="Skyview Value")

        # Plot the conic mean processed array
        plt.subplot(1, 2, 2)
        plt.imshow(skyview_conic_mean, cmap='viridis', origin='lower')
        plt.title(f"Conic Mean Processed - Zone {zone_number}")
        plt.colorbar(label="Processed Value")

        # Display the plots
        plt.tight_layout()
        plt.show()
    
    # Return the populated DataFrame
    return data_frame
def set_order(dataframe):
    return dataframe[[ 'skyview_conic_mean' ]]

# Call the function to process the zones and get the DataFrame
data_frame = load_raw_and_create_df()

# Optionally, display the first few rows of the DataFrame
print(data_frame.head())
```
