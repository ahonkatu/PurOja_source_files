### View feature images

To watch the datafiles, use this code by altering the file_path_template-name. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Change the file name accrodind the file you want to evaluate
file_path_template = "../../01_Data/01_Raw/features/features/SVF_{}.npy"  # change this name here, the files are in 01_Data/01_Raw/features/features

# This loop needs to have the amount of images, the study area squares
for i in range(11, 22):  # Numerot 1–21
    file_path = file_path_template.format(i)  # fills the filenumber
    
    try:
        # Download file
        data = np.load(file_path)
        print(f"Ladataan tiedosto: {file_path}")

        # Handle no-data values (-99999)
        # Replace no-data values with NaN for easier handling
        data[data == -99999] = np.nan

        # Is checjking out the data dimension and reading the data
        if data.ndim == 1:
            # 1D-data (for example timeline)
            plt.figure()
            plt.plot(data)
            plt.title(f"1D Data - Tiedosto {i}")
            plt.xlabel("Indeksi")
            plt.ylabel("Arvo")
        elif data.ndim == 2:
            # 2D-data (for example matrix or image)
            plt.figure()
            #plt.imshow(data, cmap='viridis', aspect='auto') #different colour option
            plt.imshow(data, cmap='gray', aspect='auto') 
            plt.colorbar(label="Arvo")
            plt.title(f"2D Data - Tiedosto {i}")
            plt.xlabel("X-akseli")
            plt.ylabel("Y-akseli")
        else:
            print(f"Datan dimensio {data.ndim} tiedostossa {file_path} ei ole tuettu.")
            continue

        # Is showing every img
        plt.show()

    except FileNotFoundError:
        print(f"Can't find the file: {file_path}")
    except Exception as e:
        print(f"Error while handling the file: {file_path}: {e}")
```
